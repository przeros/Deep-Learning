```python
import os
from PIL import Image
import glob

data_dir = '/content/dog/Dog'
# image_size = (64, 64)
image_size = (64, 64)

# List to store image data
dataset = []

# Iterate through the directory to load and preprocess images
for class_name in os.listdir(data_dir):
    class_dir = os.path.join(data_dir, class_name)
    if os.path.isdir(class_dir):
        for image_path in glob.glob(os.path.join(class_dir, '*.jpg')):
            try:
                # Load the image using PIL
                img = Image.open(image_path)
                img = img.resize(image_size)  # Resize the image
                img_array = np.array(img)  # Convert to numpy array

                # Check and enforce 3 channels (RGB)
                if len(img_array.shape) != 3 or img_array.shape[2] != 3:
                    continue

                # Append image and label
                dataset.append(img_array)

            except Exception as e:
                print(f"Error processing {image_path}: {str(e)}")

# Convert lists to numpy arrays
dataset = np.array(dataset)

# Print shape of image_data
print("Shape of image_data:", dataset.shape)

# Split the dataset into training (99%) and validation (1%) sets
split_ratio = 0.99
split_index = int(len(dataset) * split_ratio)

x_train = dataset[:split_index]
x_val = dataset[split_index:]

# Print the shapes of the training and validation datasets
print("Shape of x_train:", x_train.shape)
print("Shape of x_val:", x_val.shape)
```

### Residual Block

```python
class ResidualBlock(tf.keras.layers.Layer):
  def __init__(self, filters1, rescale=False, b_name='1'):
    super(ResidualBlock, self).__init__()
    self.rescale = rescale
    filters2 = filters1
    filters3 = 4 * filters1
    ch_axis = -1
    self.stride = 1
    if self.rescale == True:
      self.stride = 2
    w_init = 'he_uniform'
    self.base_name = b_name

    self.skip_conv1 = tf.keras.layers.Conv2D(filters1, kernel_size=1, strides=(self.stride, self.stride),
                                             kernel_initializer=w_init, name=f'{self.base_name}_skip_conv1')
    self.skip_bn1 = tf.keras.layers.BatchNormalization(axis=ch_axis, name=f'{self.base_name}_skip_bn1')
    self.conv1 = tf.keras.layers.Conv2D(filters2, kernel_size=3, padding='same', strides=(self.stride, self.stride),
                                       kernel_initializer=w_init, name=f'{self.base_name}_conv1')
    self.bn1 = tf.keras.layers.BatchNormalization(axis=ch_axis, name=f'{self.base_name}_bn1')
    self.relu1 = tf.keras.layers.Activation('relu', name=f'{self.base_name}_relu1')
    self.conv2 = tf.keras.layers.Conv2D(filters3, kernel_size=3, padding='same',
                                       kernel_initializer=w_init, name=f'{self.base_name}_conv2')
    self.bn2 = tf.keras.layers.BatchNormalization(axis=ch_axis, name=f'{self.base_name}_bn2')
    self.relu2 = tf.keras.layers.Activation('relu', name=f'{self.base_name}_relu2')

  def call(self, inputs):
    skip_conn = inputs
    if self.rescale == True:
      self.stride = 2
      skip_conn = self.skip_conv1(skip_conn)
      skip_conn = self.skip_bn1(skip_conn)

    x = self.conv1(inputs)
    x = self.bn1(x)
    x = self.relu1(x)
    x = self.conv2(x)
    x = self.bn2(x)
    x = tf.keras.layers.Add()([x, skip_conn])
    output = self.relu2(x)

    return output
```

### Architektura encodera i decodera

Przprowadzone zostało wiele testów w celu znalezienia najoptymalniejszej architektury. Eksperymenty zawierały zmiany w:


*   ilości bloków rezydualnych

*   ilość filtrów

Ostateczna architektura została zaprezentowana poniżej


```python
from keras.losses import BinaryCrossentropy, MeanSquaredError

# Parametry modelu
input_shape = (64, 64, 1)
# input_shape = (256, 256, 1)
latent_dim = 1024

class VAE(tf.keras.Model):
  def __init__(self, input_shape=input_shape, latent_dim=latent_dim):
    super(VAE, self).__init__()

    # Encoder
    self.encoder = tf.keras.Sequential([
        tf.keras.layers.InputLayer(input_shape=input_shape),
        tf.keras.layers.Conv2D(32, kernel_size=7, strides=(2, 2), activation='relu', padding='same'),

        ResidualBlock(8, b_name=f"enc_residual1"),
        # ResidualBlock(4, b_name=f"enc_residual2"),
        # ResidualBlock(4, b_name=f"enc_residual3"),

        tf.keras.layers.Conv2D(64, kernel_size=3, activation='relu', padding='same'),
        ResidualBlock(16, b_name=f"enc_residual4"),
        # ResidualBlock(8, b_name=f"enc_residual5"),
        # ResidualBlock(8, b_name=f"enc_residual6"),
        # ResidualBlock(8, b_name=f"enc_residual7"),
        tf.keras.layers.MaxPooling2D((2, 2), padding='same'),

        tf.keras.layers.Conv2D(128, kernel_size=3, activation='relu', padding='same'),
        ResidualBlock(32, b_name=f"enc_residual8"),
        # ResidualBlock(16, b_name=f"enc_residual9"),
        # ResidualBlock(16, b_name=f"enc_residual10"),
        # ResidualBlock(16, b_name=f"enc_residual11"),
        # ResidualBlock(16, b_name=f"enc_residual12"),
        # ResidualBlock(16, b_name=f"enc_residual13"),
        tf.keras.layers.MaxPooling2D((2, 2), padding='same'),

        tf.keras.layers.Conv2D(256, kernel_size=3, activation='relu', padding='same'),
        ResidualBlock(64, b_name=f"enc_residual14"),
        # ResidualBlock(32, b_name=f"enc_residual15"),
        # ResidualBlock(32, b_name=f"enc_residual16"),

        tf.keras.layers.MaxPooling2D((2, 2), padding='same'),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(latent_dim)
    ])
    self.log_var = tf.keras.layers.Dense(latent_dim)
    self.sampling = Sampling()

    # Decoder
    self.decoder = tf.keras.Sequential([
        tf.keras.layers.InputLayer(input_shape=(latent_dim,)),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dense(8 * 8 * 256, activation='relu'),
        tf.keras.layers.Reshape((8, 8, 256)),
        tf.keras.layers.UpSampling2D((2, 2)),

        ResidualBlock(64, b_name=f"enc_residual1"),
        # ResidualBlock(32, b_name=f"enc_residual2"),
        # ResidualBlock(32, b_name=f"enc_residual3"),
        tf.keras.layers.UpSampling2D((2, 2)),

        tf.keras.layers.Conv2D(128, kernel_size=3, activation='relu', padding='same'),
        ResidualBlock(32, b_name=f"enc_residual4"),
        # ResidualBlock(16, b_name=f"enc_residual5"),
        # ResidualBlock(16, b_name=f"enc_residual6"),
        # ResidualBlock(16, b_name=f"enc_residual7"),
        # ResidualBlock(16, b_name=f"enc_residual8"),
        # ResidualBlock(16, b_name=f"enc_residual9"),
        tf.keras.layers.UpSampling2D((2, 2)),

        tf.keras.layers.Conv2D(64, kernel_size=3, activation='relu', padding='same'),
        ResidualBlock(16, b_name=f"enc_residual8"),
        # ResidualBlock(8, b_name=f"enc_residual11"),
        # ResidualBlock(8, b_name=f"enc_residual12"),
        # ResidualBlock(8, b_name=f"enc_residual13"),

        tf.keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding='same'),
        ResidualBlock(8, b_name=f"enc_residual14"),
        # ResidualBlock(4, b_name=f"enc_residual15"),
        # ResidualBlock(4, b_name=f"enc_residual16"),

        tf.keras.layers.Conv2D(1, kernel_size=3, activation='sigmoid', padding='same')
    ])

  def encode(self, x):
    mean = self.encoder(x)
    log_var = self.log_var(mean)
    z = self.sampling([mean, log_var])
    return z, mean, log_var

  def decode(self, z):
    return self.decoder(z)

  def call(self, inputs):
    x = inputs
    z, mean, log_var = self.encode(x)
    decoded_output = self.decode(z)
    return decoded_output

class Sampling(tf.keras.layers.Layer):
  def call(self, inputs):
    mean, log_var = inputs
    epsilon = tf.keras.backend.random_normal(shape=(tf.shape(mean)[0], tf.shape(mean)[1]), mean=0.0, stddev=1.0)
    return mean + tf.exp(0.5 * log_var) * epsilon

beta = 0.0000001
beta_decay = 0.00001
beta_final = 0.001

@tf.function
def train_step(model, x, optimizer):
  with tf.GradientTape() as tape:
    # Encoder
    z, mean, log_var = model.encode(x)
    # Decoder
    x_decoded = model.decode(z)
    loss, reconstruction_loss, kl_loss = vae_loss_function(x, x_decoded, mean, log_var)
  gradients = tape.gradient(loss, model.trainable_variables)
  optimizer.apply_gradients(zip(gradients, model.trainable_variables))
  return {'loss': loss, 'kl_loss': kl_loss, 'reconstruction_loss': reconstruction_loss}

def vae_loss_function(x, x_decoded, mean, log_var):
  reconstruction_loss = MeanSquaredError()(x, x_decoded)
  kl_loss = -0.5 * tf.reduce_sum(1 + log_var - tf.square(mean) - tf.exp(log_var), axis=-1)
  return reconstruction_loss + beta * kl_loss, reconstruction_loss, kl_loss
```

### Trening modelu

Optymalizowane były hiperparametry. Ilość epok będzie wynosić około 100 (żeby zmieścić się w przedziale 4h treningu). Batch size dawał najlepsze rezultaty dla wartości 128. Learning rate ustawiony jest na 0.001 z włączonym exponentialDecayRate = 0.99 oraz posiadamy **decay** dla wartości parametru **beta początkowa = 0.00001 oraz decay ustawiony na wartość = 0.0000001**

Co piątą epokę treningu robimy przykładowe rekonstrukcje na zdjęciach x_val aby zobaczyć jak na danym etapie model sobie radzi.

```python
import time

start_time = time.time()


vae = None
vae = VAE(input_shape=input_shape, latent_dim=latent_dim)
initial_learning_rate = 0.001
final_learning_rate = 0.0001
epochs = 120
batch_size = 128
num_samples = x_train.shape[0]

lr_schedule = keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate,
    decay_steps=epochs,
    decay_rate=0.99,

)

optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)

# Build the model to ensure proper initialization
vae.build((batch_size, *input_shape))
vae.summary()

# Initialize the current learning rate variable
current_learning_rate = initial_learning_rate

# Iterate through epochs
start_time = time.time()
for epoch in range(1, epochs + 1):
    print(f'epoch: {epoch}')

    total_loss = 0.0  # Initialize total loss for the epoch
    total_reconstruction_loss = 0.0
    total_kl_loss = 0.0

    # Shuffle and iterate through batches
    indices = np.arange(num_samples)
    np.random.shuffle(indices)

    for i in range(0, num_samples - (batch_size - 1), batch_size):
        batch_indices = indices[i:i+batch_size]
        train_batch = x_train[batch_indices]

        #Apply data augmentation to the batch
        augmented_images = []
        for image in train_batch:
            augmented_image1 = data_generator.random_transform(image)
            augmented_image2 = data_generator.random_transform(image)
            augmented_image3 = data_generator.random_transform(image)
            augmented_image4 = data_generator.random_transform(image)

            augmented_images.extend([augmented_image1, augmented_image2, augmented_image3, augmented_image4])

        augmented_images = np.array(augmented_images)
        train_batch = np.vstack([train_batch, augmented_images])

        loss = train_step(vae, train_batch, optimizer)
        total_reconstruction_loss += loss['reconstruction_loss'].numpy()
        total_kl_loss += loss['kl_loss'].numpy()
        total_loss += loss['loss'].numpy()

    # Update the current learning rate
    current_learning_rate = optimizer.lr.numpy()
    # Compute the mean loss for the epoch
    mean_loss = tf.reduce_mean(total_loss)
    total_time = time.time() - start_time
    print(f'Epoch {epoch} - Mean Loss: {mean_loss} - KL_loss: {tf.reduce_mean(total_kl_loss)} - reconstruction_loss: {tf.reduce_mean(total_reconstruction_loss)} - learning_rate: {current_learning_rate} total_time:{total_time}')
    beta = min(beta_final, beta + beta_decay)
    if total_time > 14280:
      reconstruct_images(vae, x_val[epoch - 10:epoch], n=10)
      generate_images(vae, (10, 1024), n=10)
      break

    # Test
    if (epoch % 10 == 0):
      reconstruct_images(vae, x_val[epoch - 10:epoch], n=10)
      generate_images(vae, (10, 1024), n=10)
```
